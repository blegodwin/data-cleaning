# Data Transformation

- round up float column to two decimal places

- truncate values. separate numeric data from text.

```python
arewa['GPA'].info
```

Notice how the column type is  object although it seemingly contain floats?

```python
# convert column type from object to float
arewa[['GPA']] = arewa[['GPA']].astype(float)
```

The error message above shows Type Error. The column contains the str value 'personal'

```python
# the functioncconverts the column to numeric type and invalid entries are converted to NaN
arewa[['GPA']] = arewa[['GPA']].apply(pd.to_numeric, errors = 'coerce')
```

```python
# the numeric float values are rounded up to two decimal places
arewa['GPA'] = arewa['GPA'].round(2)
```

```python
arewa
```

##### Separate numeric value attached to string value 

```python
arewa['Weight'].isnull().any()
```

```python
# using the "Weight" column
arewa['Weight'] = arewa['Weight'].str.replace(r'\D', '', regex = True).replace('', np.nan)
```

```python
arewa['Weight']
```

In the 'Weight' column, the function and it's positional args remove all non-numeric characters and values that are just string types to NaN. This column is still an object type and can be converted to int.

```python
arewa['Weight'] = arewa['Weight'].apply(pd.to_numeric, errors = 'coerce').astype
```
